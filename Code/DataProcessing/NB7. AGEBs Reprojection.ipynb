## 2. Sets of AGEBs: All, Same, Divided, New

````python
# All AGEBs
all_1990 = set(census_1990.CVEGEO)
all_2000 = set(census_2000.CVEGEO)
all_2010 = set(census_2010.CVEGEO)
all_2020 = set(agebs_2020.CVEGEO)

# AGEBs which remains between census
same_1990_2000 = all_1990.intersection(all_2000)
same_2000_2010 = all_2000.intersection(all_2010)
same_2010_2020 = all_2010.intersection(all_2020)

# The divided or dissapearing agebs are those in the past that dont survived
divided_1990 = all_1990 - same_1990_2000
divided_2000 = all_2000 - same_2000_2010
divided_2010 = all_2010 - same_2010_2020

# New AGEBs can come from divided or completely new AGEBs.
new_2000 = all_2000 - same_1990_2000
new_2010 = all_2010 - same_2000_2010
new_2020 = all_2020 - same_2010_2020
````

## 3. AGEBs Reprojection

````python
# Function to filter duplcated target agebs
def rank_max(s):
    m = s.max()
    return s == m

# Function to exclude agebs that would exceed original area by margin
def limit_area(df):
    # Sort by shared area, we proposed new agebs based on the amount of overlap
    shared_area = df.shared_area.values
    areas_idx = np.argsort(-shared_area)
    
    # We accumulate the new agebs total area / old ageb area (fraction area)
    frac_area = df.frac_area.values[areas_idx]
    # Deviations from unity
    cum_area = np.cumsum(frac_area)
    dev = abs(cum_area - 1.0)
    # Find index closest to unity
    max_idx = np.argmin(dev)

    # First first index that exceeds area
    # mask = np.cumsum(frac_area) < 1.01
    # Edge cases
    # if np.all(mask):
    #     # All less, take all agebs
    #     max_idx = len(df)
    # elif np.all(~mask):
    #     # First one exceeds, take 1
    #     max_idx = 1
    # else:
    #     max_idx = np.argmin(mask)

    valid_idx = np.zeros_like(shared_area, dtype=bool)
    valid_idx[areas_idx[:max_idx+1]] = True
    #return pd.DataFrame({'valid': valid_idx, 'CVEGEO_2000': df.CVEGEO_2000})
    return pd.DataFrame({'valid': valid_idx})
````

### 3.1 1990 - 2000

#### 3.1.1 Divided AGEBs

````python
# Area overlap of divided and new
census_1990_divided = census_1990[census_1990.CVEGEO.isin(divided_1990)]
census_2000_new = census_2000[census_2000.CVEGEO.isin(new_2000)]
print("Completado: AGEBs divididas y nuevas")

overlap_1990_2000 = census_1990_divided.sjoin(
    census_2000_new.assign(geometry_2000=census_2000_new.geometry),
    how='left',
    lsuffix='1990',
    rsuffix='2000')

print("Completado: AGEBs traslapadas")

overlap_1990_2000['shared_area'] = overlap_1990_2000.geometry.intersection(overlap_1990_2000.geometry_2000).area / overlap_1990_2000.Area_1990
overlap_1990_2000['frac_area'] = overlap_1990_2000.Area_2000 / overlap_1990_2000.Area_1990

print("Completado: Obtener shared_area y frac_area")

# Target ageb should only appear once
overlap_1990_2000 = overlap_1990_2000[overlap_1990_2000.groupby('CVEGEO_2000')['shared_area'].transform(rank_max)]

print("Completado: Solo debe aparecer un AGEB objetivo")

# Check centroids of new agebs inside old ones
overlap_1990_2000['centroids_2000'] = overlap_1990_2000.geometry_2000.centroid
overlap_1990_2000['centroid_inside'] = overlap_1990_2000.centroids_2000.within(overlap_1990_2000.geometry)
overlap_1990_2000 = overlap_1990_2000[overlap_1990_2000.centroid_inside]

# Total new area should not exceed original area
# se agregó un .values por que marcaba un error por el indexado
overlap_1990_2000 = overlap_1990_2000[overlap_1990_2000.groupby('CVEGEO_1990').apply(limit_area)['valid'].values]

fig, ax = plt.subplots(figsize=(50,50))
census_1990_divided.plot(ax=ax, facecolor='grey', edgecolor='red', alpha=0.5, lw=3)
overlap_1990_2000.geometry_2000.plot(ax=ax, facecolor='none', edgecolor='blue')
for s, c in zip(census_1990_divided.CVEGEO, census_1990_divided.centroid):
    ax.text(c.xy[0][0], c.xy[1][0], s);
for s, c in zip(overlap_1990_2000.CVEGEO_2000, overlap_1990_2000.geometry_2000.centroid):
    ax.text(c.xy[0][0], c.xy[1][0], s, color='blue');
overlap_1990_2000.geometry_2000.centroid.plot(ax=ax)
````

```python
def reproject_population(data, CVEGEO_column, pobtot_column, frac_area_column, pobtot_reprojected_column):
    df = data.copy()
    # Frac Area Normalization
    df[frac_area_column] = df[frac_area_column].div(df.groupby([CVEGEO_column])[frac_area_column].transform('sum'))
    
    # New column with "Reprojected Population"
    df[pobtot_reprojected_column] = (df[pobtot_column] * df[frac_area_column]).apply(np.floor)
    
    # Making corrections on "Reprojected Population"
    # Add to AGEB with maximum frac area per group the difference between 
    # pobtot_column (per group) and sum of pobtot_reprojected (per group)

    # Steps:
    # 1. Column with reprojected population
    # 2. Create a column with the difference between poptot_column and the sum of pobtot_reprojected per group. 
    #    It will result with same value for each element of group.
    # 3. Multiply previous column with boolean column (0 & 1s) with 1 where the maximum element for frac_area per group
    
    df[pobtot_reprojected_column] = df[pobtot_reprojected_column] +\
                               (df[pobtot_column] - df.groupby([CVEGEO_column])[pobtot_reprojected_column].transform('sum')) * \
                               df.groupby(CVEGEO_column)[frac_area_column].transform('max').eq(df[frac_area_column]) 
    return df
```

````python
def reproject_population(data, CVEGEO_column, pobtot_column, frac_area_column, pobtot_reprojected_column):
    df = data.copy()
    # Frac Area Normalization
    df[frac_area_column] = df[frac_area_column].div(df.groupby([CVEGEO_column])[frac_area_column].transform('sum'))
    
    # New column with "Reprojected Population"
    df[pobtot_reprojected_column] = (df[pobtot_column] * df[frac_area_column]).apply(np.floor)
    
    # Making corrections on "Reprojected Population"
    # Add to AGEB with maximum frac area per group the difference between 
    # pobtot_column (per group) and sum of pobtot_reprojected (per group)

    # Steps:
    # 1. Column with reprojected population
    # 2. Create a column with the difference between poptot_column and the sum of pobtot_reprojected per group. 
    #    It will result with same value for each element of group.
    # 3. Multiply previous column with boolean column (0 & 1s) with 1 where the maximum element for frac_area per group
    
    df[pobtot_reprojected_column] = df[pobtot_reprojected_column] +\
                               (df[pobtot_column] - df.groupby([CVEGEO_column])[pobtot_reprojected_column].transform('sum')) * \
                               df.groupby(CVEGEO_column)[frac_area_column].transform('max').eq(df[frac_area_column]) 
    return df
````
